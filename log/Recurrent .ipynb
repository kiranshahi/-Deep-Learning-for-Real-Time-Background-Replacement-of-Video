```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
log = pd.read_csv('csv/Set1_resnet.csv')
```

```python
def plot_digram(data, lim, xlabel, ylabel):
    for current_data in data:
        plt.plot(log[current_data['key']], label=current_data['label'])
    plt.ylim(lim)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True)
```

### Human centric video matting

```python
data = [{'label': 'Loss', 'key':'loss'}, {'label': 'Validation Loss', 'key':'val_loss'}]
plot_digram(data, [0, 0.6], 'Epoch', 'Dice Loss')
```

```python
data = [{'label': 'MSE', 'key':'mean_squared_error'}, {'label': 'Validation MSE', 'key':'val_mean_squared_error'}]
plot_digram(data, [0, 0.3], 'Epoch', 'Mean Squared Error')
```

```python
data = [{'label': 'Learning Rate', 'key':'lr'}]
plot_digram(data, [0, 0.00012], 'Epoch', 'Learning Rate')
```
