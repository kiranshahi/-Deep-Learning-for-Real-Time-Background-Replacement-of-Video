```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
log = pd.read_csv('csv/mobilenetv3_unet_single_frame.csv')
```

```python
log
```

```python
plt.plot(log['loss'], label='loss')
plt.plot(log['val_loss'], label='val_loss')
plt.ylim([0, 0.3])
plt.xlabel('Epoch')
plt.ylabel('Dice loss')
plt.legend()
plt.grid(True)
```

```python
plt.plot(log['mean_squared_error'], label='MSE')
plt.plot(log['val_mean_squared_error'], label='Val MSE')
plt.ylim([0, 0.3])
plt.xlabel('Epoch')
plt.ylabel('Mean Squared Error')
plt.legend()
plt.grid(True)
```

```python
plt.plot(log['lr'], label='Learning Rate')
plt.ylim([0, 0.00012])
plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
plt.legend()
plt.grid(True)
```
